# Algoritmi Genetici: Risoluzione del problema del commesso viaggiatore

Lo scopo di questa esercitazione è di studiare e applicare algoritmi genetici alla risoluzione del problema del commesso viaggiatore. Il problema può essere schematizzato come segue:
- Un certo numero di città sono distribuite, uniformemente o secondo uno schema, in un piano (nel nostro caso) 2-dimensionale
- Un viaggiatore (o venditore) deve visitare tutte le città una e una sola volta, ritornando infanie al punto di partenza
Ci si pone come obiettivo di trovare la strada (che è essenzialmente caratterizzata dalla permutazione che indica l'ordine in cui si visitano le città) che sia più economica (e quindi più corta). Utilizzando la terminologia e la schematizzazione di un problema di ottimizzazione, il problema si traduce nel trovare la permutazione che minimizza la funzione di costo 

$$
L^{(1)}(\vec{x_1},\dots,\vec{x_n}) = \sum_i |\vec{x_{i+1}}-\vec{x_i}|
$$
(che corrisponde alla lunghezza del percorso) o equivalentemente di $L^{(2)}(\vec{x_1},\dots,\vec{x_n}) = \sum_i |\vec{x_{i+1}}-\vec{x_i}|^2 $.

Gli algoritmi genetici sono una classe di algoritmi che si propongono di risolvere problemi come questo traendo ispirazione dai meccanismi di selezione naturale. L'idea è di osservare lo sviluppo di una popolazione di "cromosomi" che presentino mutazioni genetiche spontanee con una certa probabilità $p_m$ e che si riproducano, mescolando il patrimonio genetico, con una certa probabilità $p_c$. In analogia al mondo naturale, i cromosomi migliori (quelli con costi dati da $L^{(1)}$ o $L^{(2)}$ minori) vengono scelti con una probabilità maggiore per la riproduzione di quelli con costi elevati.

#### Implementazione

Per affrontare questo problema sono state sviluppate alcune classi e funzioni, contenute nella libreria "tsp.h".
Le possibili alternative nel tragitto sono state tradotte in vettori di interi, che rappresentano la sequenza degli indici delle città nell'ordine specifico di visita. Il primo elemento di ogni vettore è 0, e ogni indice deve apparire una e una sola volta. La funzione booleana TSPCheckPerm(vector<int>) controlla se la permutazione è ammissibile.
Le classi sviluppate sono:
- **Classe TSP**: rappresenta un generico problema del commmesso viaggiatore. È caratterizzata da un vettore di punti 2-dimensionali (che rappresentano le posizioni delle città), da funzioni membro per generare diverse distribuzioni di città (uniforme o su di una circonferenza) e da una funzione membro che calcola il costo di una data permutazione (nella forma di  $L^{(2)}$).
- **Classe GeneticTSP**: è un oggetto per applicare un algoritmo genetico a un oggetto TSP (da cui eredita). È dotata di due metodi di mutazione, che scambiano due indici oppure che spostano di un certo valore ciascun indice del vettore (tranne il primo), uno di crossover (riproduzione) che combina i due genitori in due figli e uno di selezione (la selezione è più probabile quanto più basso è il costo). Sono presenti delle funzioni di analisi dei costi, un costruttore che genera una popolazione con variazioni ottenute mutando un primo vettore e una funzione membro MakeGeneration che permette di sviluppare una nuova generazione tramite gli operatori di selezione, crossover e mutazione. Il data membro principale, che rappresenta la popolazione, è un vettore di vettori di interi.
- **Classe Sorter**: è un funtore per applicare gli algoritmi della libreria standard (std::sort()) alla popolazione e ordinarla in funzione del costo di una permutazione.


#### Efficienza dell'operatore di selezione



```python
#view how the selection operator works

import matplotlib.pyplot as plt
import numpy as np

h = np.loadtxt('output/selection_operator/selection.dat')
plt.hist(h,density=True,label='selections/total')
x,y = np.loadtxt('output/selection_operator/prob.dat',unpack=True)
plt.plot(x,y,'r^',label='probability to be selected')
plt.grid(True)
plt.xlabel('cost')
plt.ylabel('probability')

plt.legend()
plt.show()
```

Per avere un'idea di come e quanto vengono selezionati i vari elementi della popolazione si può osservare questo istogramma: rappresenta quanto spesso viene selezionato un elemento dato il suo costo. Il test è stato effettuato su 1000 selezioni.

La probabilità di selezione è data da $p(cost) = \big(\frac{cost-cost_{max}}{cost_{min}-cost_{max}}\big)^p $, con $p=3$. Si può variare l'esponente per ottenere un operatore di selezione più o meno rigoroso.

Parametri della simulazione:
- numero di generazioni: 500
- elementi della popolazione: 500
- $p_{crossover} = 0.7$, $p_{mutation} = 0.1$

## Problema del commesso viaggiatore: 32 città distribuite su di una circonferenza

In questa sezione si analizzano i risultati dell'applicazione delle tecniche sopra discusse a un problema in cui 32 città sono distribuite casualmente su di una circonferenza. In questa particolare conformazione non è difficile capire che la soluzione più efficiente è quella di visitare le città in ordine, percorrendo mano a mano l'intera circonferenza. Avere una "soluzione analitica" è in questo caso molto utile per confrontarsi con i risultati ottenuti con l'algoritmo e capirne l'efficienza e le potenzialità.

```python
#see how the lowest cost goes

import matplotlib.pyplot as plt
import numpy as np

y = np.loadtxt('output/circle/lowest.dat')
x = np.arange(y.size)
means, err = np.loadtxt('output/circle/lowerhalf.dat',unpack=True)
blocks = 20*np.arange(means.size)


plt.figure(figsize=(10,7))
plt.plot(x,y,'r',label='best path cost')
plt.errorbar(blocks,means,err,label='best half of the population mean cost')
plt.grid(True)
plt.xlabel('generations')
plt.ylabel('L2 cost')
plt.title("Cost of the best & best half of the combinations - Genetic Algorithm TSP",fontweight='bold')
plt.legend()
plt.show()
```

In questo grafico sono rappresentati il costo della permutazione migliore e, analizzate con la tecnica del Data Blocking, il costo medio della metà migliore di popolazione (ogni media è un elemento del blocco). L'algoritmo sembra essere decisamente efficiente nella sua ricerca.

### Visualizzazione del percorso migliore

```python
import matplotlib.pyplot as plt
import numpy

x,y = np.loadtxt("output/circle/bestpath.dat", unpack=True)
x = numpy.append(x, x[0])
y = numpy.append(y, y[0])

plt.figure(figsize=(8,8))
plt.plot(x,y)
plt.plot(x,y,'r^',label='cities')
plt.grid(True)
plt.title('Cities distribution & best path',fontweight='bold')

plt.legend()
plt.show()
```

La risposta finale dell'algoritmo non è quella ottimale, ma vi si avvicina.
