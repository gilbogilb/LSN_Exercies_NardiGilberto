```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)
```

```python
#loading dataset

from PIL import Image
import os

data_list = []
for i in range(10):
    digit_filename = 'img/12.3/'+str(i)+'.png'
    digit_in = Image.open(digit_filename).convert('L')
    ydim, xdim = digit_in.size
    #print("Image size: "+str(xdim)+"x"+str(ydim))
    pix=digit_in.load();
    data = np.zeros((xdim, ydim))
    for j in range(ydim):
        for k in range(xdim):
            data[k,j]=pix[j,k]
    data/= 255
    data_list.append(data)


test = np.array(data_list)  #converting the list in a np array
test=test.reshape(10,xdim,ydim,1) #reshaping it to fit the input type for the DCNN model

plt.figure(figsize=(15,4))
for i in range(10):
    plt.subplot(1, 10, i + 1)
    plt.imshow(test[i, :, :, 0], cmap='gray')
plt.show()
```

```python
#loading DCNN model (Ex12.2)
save_model_path='DCNN_model'
model = tf.keras.models.load_model(filepath=save_model_path)
model.summary()
```

```python
#prediction
predictions = model.predict(test)
target = np.arange(10)

#img_rows, img_cols = 28
##test = test.reshape(test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(17, 5)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(test[i, :, :, 0], cmap='gray')    
    plt.title( 'predicted as '+str(np.argmax(predictions[i])))   
    plt.axis('off') 
plt.show()
```
