# Simulazioni di dinamica molecolare
In questa esercitazione si migliora e utilizza il codice fornito per simulare l'evoluzione di un sistema di molecole (monoatomiche) che interagiscono secondo il potenziale di Lennard-Jones
$$V(r)=4\epsilon \big[\big(\frac{\sigma}{r}\big)^{12}-\big(\frac{\sigma}{r}\big)^6 \big]$$
Il modello scelto è quello relativo a un gas di Argon: $\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu

## Esercizio 04.1
La prima operazione effettuata è stata apportare le modifiche suggerite al codice. In particolare, abbiamo definito:

- Una variabile <font face="Courier">bool restart</font> che permette di decidere se fare cominciare la simulazione da zero o se farla ripartire da configurazioni successive note. Questa differenza è importante perché il codice implementa l'algoritmo di Verlet per integrare numericamente le equazioni del moto. Questo algoritmo è un metodo two-step: si ha che la posizione $\vec{r}(t+dt)$ dipende dalle posizioni al tempo $t$ e $t-dt$, quindi non basta conoscere posizione e velocità iniziali delle molecole per cominciare la simulazione. Nel caso in cui il programma sia eseguito **per la prima volta**, si sceglie <font face="Courier">restart=false</font>: il programma parte da una configurazione salvata in <font face="Courier">inputs/config.0</font> e genera delle velocità casuali (con velocità del centro di massa nulla) $\vec{v}(t_0+dt/2)$. La configurazione di partenza è importante nelle simulazioni di stato solido (abbiamo scelto una configurazione di cristallo fcc), ma meno determinante per le fasi liquida e gassosa: si ha infatti che dopo pochi passi di simulazione il cristallo "si scioglie" e le molecole iniziano i moti disordinati caratteristici di queste fasi. Se invece si esegue il programma **riprendendo una simulazione precedente**, scegliendo <font face="Courier">restart=true</font> si possono sfruttare le integrazioni già eseguite per delle stime di $\vec{r}(t_0)$ e $\vec{r}(t_0-dt)$, prodotte in output dalla simulazione precedente e lette dal programma rispettivamente in <font face="Courier">inputs/config.0</font> e <font face="Courier">inputs/config.old</font>. A questo punto il programma esegue una volta l'algoritmo di Verlet, ottenendo $\vec{r}(t+dt)$, e calcola le velocità $\vec{v}(t_0+dt/2)$.
- Una variabile <font face="Courier">bool rescale</font> che permette di decidere se riscalare o meno le velocità all'inizio dell'esecuzione. Si ha infatti che le velocità iniziali, se il sistema non è equilibrato, non sono proprie di un sistema alla temperatura desiderata (che avrebbe velocità distribuite secondo la distribuzione di Maxwell-Boltzmann): bisogna quindi calcolare la temperatura effettiva del sistema, confrontarla con la temperatura desiderata e ottenere un fattore di scala per le velocità. Dopodiché di possono correggere le posizioni iniziali $\vec{r}_{new}(t_0)=\vec{r}(t_0+dt)-\vec{v}_s dt$ (dove $v_s$ sono le velocità riscalate) e iniziare la simulazione con $\vec{r}_{new}(t_0)$, $\vec{r}(t_0+dt)$.

#### Equilibrazione delle fasi
Con l'implementazione delle modifiche sopra descritte, iterando l'esecuzione del programma facendo ripartire la simulazione dalle configurazioni precedenti (e riscalando le velocità) si riesce a equilibrare il sistema. Concentriamo la nostra attenzione sulla temperatura: si osserva che se durante le prime simulazioni il suo valore si discosta significativamente dal valore richiesto, durante le successive iterazioni ci si avvicina via via di più fino ad ottenere configurazioni la cui temperatura oscilla stabilmente attorno quella desiderata.

Mostriamo l'evoluzione della temperatura del sistema in funzione delle misure, effettuate una volta ogni $10$ passi di simulazione, nelle seguenti condizioni termodinamiche:
1. **fase solida**: $\rho^\star = 1.1$, $T^\star = 0.8$
2. **fase liquida**: $\rho^\star = 0.8$, $T^\star = 1.1$
3. **fase gassosa**: $\rho^\star = 0.05$, $T^\star = 1.2$

In tutti i casi è evidenziata la temperatura richiesta $T_0$ (in unità di Lennard Jones). La ripartenza della simulazione è effettuata ogni $10^3$ misure.

```python
import matplotlib.pyplot as plt
import numpy as np

#loading data
solid   = np.loadtxt('outputs/equilibration/Tsolid.dat')
xsolid  = np.arange(solid.size)
T0solid = 0.8

liquid = np.loadtxt('outputs/equilibration/Tliquid.dat')
xliquid = np.arange(liquid.size)
T0liquid= 1.1 

gas   = np.loadtxt('outputs/equilibration/Tgas.dat')
xgas  = np.arange(gas.size)
T0gas = 1.2

#graphing
plt.figure(figsize=(15,18))
plt.suptitle('Equilibrazione',fontweight='bold', fontsize = 16)

plt.subplot(311)
plt.plot(xsolid,solid,label='Temperatura')
plt.plot(xsolid,T0solid*np.ones(solid.size),'r--',label='$T_0$')
plt.grid(True)
plt.title('Fase solida',fontsize=13)
plt.ylabel('$T^*$')
plt.xlabel('misure')
plt.legend()

plt.subplot(312)
plt.plot(xliquid,liquid,label='Temperatura')
plt.plot(xliquid,T0liquid*np.ones(liquid.size),'r--',label='$T_0$')
plt.grid(True)
plt.title('Fase liquida',fontsize=13)
plt.ylabel('$T^*$')
plt.xlabel('misure')
plt.legend()

plt.subplot(313)
plt.plot(xgas,gas,label='Temperatura')
plt.plot(xgas,T0gas*np.ones(gas.size),'r--',label='$T_0$')
plt.grid(True)
plt.title('Fase gassosa',fontsize=13)
plt.ylabel('$T^*$')
plt.xlabel('misure')
plt.legend()

plt.show()
```

Le fasi solida e liquida si stabilizzano piuttosto rapidamente, dopo qualche ripartenza. Quella gassosa fa più fatica a stabilizzarsi, anche dopo un grande ($>20$) numero di simulazioni. Si hanno tuttavia oscillazioni attorno a $T_0$ già dal secondo esperimento.


## Esercizio 4.2
In questa parte dell'esercitazione abbiamo esteso il codice aggiungendo la possibilità di ottenere medie e incertezze statistiche sulle misure effettuate sul sistema con la tecnica del Data Blocking. Questa funzionalità è resa possibile tramite l'implementazione di una classe <font face="Courier">DataBlockingTool</font>, che prende come argomento del costruttore la dimensione dei blocchi <font face="Courier">block_size</font> e il nome di un file in cui scrivere i risultati (media progressiva e incertezza sulla media) e gestisce automaticamente le medie a blocchi, richiedendo di volta in volta solo il valore della nuova misura.

## Esercizio 4.3
Una volta prodotto il codice con le funzionalità necessarie, abbiamo eseguito simulazioni di diverse fasi termodinamiche (solida, liquida e gassosa, con gli stessi parametri utilizzati durante l'equilibrazione). Per ogni simulazione, abbiamo equilibrato il sistema e, durante l'ultima simulazione, ne abbiamo misurato le proprietà (energia cinetica, potenziale e totale per particella, temperatura) e calcolato medie e incertezze statistiche.

In tutti i casi le grandezze sono espresse in unità SI. Abbiamo diviso i dati ottenuti in $B=50$ blocchi.

```python
#conversione unità di misura

Tscala=120.
Escala=120.*1.381E-23
```

### Fase solida

I parametri della simulazione in questa fase sono $\rho^\star = 1.1$, $T^\star = 0.8$. Il raggio di cut-off scelto è $r_c = 2.2$. Sono stati prodotti $10^4$ dati (con $10^5$ step di simulazione e una misura ogni $10$ step).

```python
K, errK = Escala*np.loadtxt('outputs/solid/ekin_blocks.dat',unpack=True,usecols=(0,1))
T, errT = Tscala*np.loadtxt('outputs/solid/temp_blocks.dat',unpack=True,usecols=(0,1))
V, errV = Escala*np.loadtxt('outputs/solid/epot_blocks.dat',unpack=True,usecols=(0,1))
E, errE = Escala*np.loadtxt('outputs/solid/etot_blocks.dat',unpack=True,usecols=(0,1))

passi = np.arange(K.size)

plt.figure(figsize=(16,12))
plt.suptitle('Fase solida',fontweight='bold', fontsize=18)

plt.subplot(221)
plt.errorbar(passi,K,errK)
plt.title('Energia cinetica per particella')
plt.xlabel('blocchi')
plt.ylabel('E/N [J]')
plt.grid(True)

plt.subplot(222)
plt.errorbar(passi,V,errV)
plt.title('Energia potenziale per particella')
plt.xlabel('blocchi')
plt.ylabel('E/N [J]')
plt.grid(True)

plt.subplot(223)
plt.errorbar(passi,E,errE)
plt.title('Energia totale per particella')
plt.xlabel('blocchi')
plt.ylabel('E/N [J]')
plt.grid(True)

plt.subplot(224)
plt.errorbar(passi,T,errT)
plt.title('Temperatura')
plt.xlabel('blocchi')
plt.ylabel('T [K]')
plt.grid(True)

plt.show()

print('---Stime finali---')
print('Energia cinetica   ',K[K.size-1],'±',errK[K.size-1],' J')
print('Energia potenziale ',V[V.size-1],'±',errV[V.size-1],' J')
print('Energia totale     ',E[E.size-1],'±',errE[E.size-1],' J')
print('Temperatura        ',T[T.size-1],'±',errT[T.size-1],' K')
```
