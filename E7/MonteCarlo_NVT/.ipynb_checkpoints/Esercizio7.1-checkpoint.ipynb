Simulazioni di fase solida, liquida e gassosa.

equilibrazione pressoche istantanea (rispetto a $5 \times 10^5$ steps)

pressione negativa all'inizio della fase liquida...


```python
import matplotlib.pyplot as plt
import numpy as np

#autocorrelation function
        
def ACF(x, tn):
    ac = np.zeros(tn)
    tmax = x.size
    for t in range(tn):
        ac[t] = (np.sum(x[0:tmax-t]*x[t:])/(tmax-t) - (np.sum(x[0:tmax - t])*np.sum(x[t:]))/(tmax-t)**2)/(np.sum(x*x)/tmax - (np.sum(x)/tmax)**2)
    return ac
```

### Fase solida
parametri soliti

```python
U, P = np.loadtxt("output/solid/solidraw.dat",unpack=True,usecols=(0,1))
steps = np.arange(U.size)

plt.figure(figsize=(16,10))
plt.suptitle('Solid phase data',fontweight='bold')


plt.subplot(211)
plt.plot(steps,U)
plt.title('Internal energy per particle')
plt.xlabel('steps')
plt.ylabel('U/N')
plt.grid(True)

plt.subplot(212)
plt.plot(steps,P)
plt.title('Pressure')
plt.xlabel('steps')
plt.ylabel('P')
plt.grid(True)

plt.show()
```

Calcoliamo l'autocorrelazione (saltiamo i primi elementi che sono di equilibrazione)

```python
#autocorrelation

eqsteps = 200
maxstep = 5000
dsteps  = maxstep-eqsteps

Usample = U[eqsteps:maxstep]
Psample = P[eqsteps:maxstep]

ACU = ACF(U,dsteps)
ACP = ACF(P,dsteps)

steps = eqsteps + np.arange(dsteps)

print(ACU.size,ACP.size)

plt.figure(figsize=(18,8))
plt.subplot(121)
plt.plot(steps,ACU)
plt.title('AC for internal energy values')
plt.xlabel('steps')
plt.ylabel('$AC_U(n)$')
plt.grid(True)

plt.subplot(122)
plt.plot(steps,ACP)
plt.title('AC for pressure values')
plt.xlabel('steps')
plt.ylabel('$AC_P(n)$')
plt.grid(True)

plt.show()
```

## Stima grandezza blocchi - Errore in funzione della dimensione del blocco
Studiamo l'errore statistico che si ottiene da questi dati in funzione della dimensione dei blocchi che scegliamo di utilizzare per il Data Blocking.

```python
class DataBlockingTool:
    
    def __init__(self,blocksize):
        self.blk_size=blocksize
        self.blk_sum=0.     #sum of the elements of this block
        self.blk_num=0      #number of blocks
        self.err=0.         #statistical error
        self.progmean=0.    #progressive mean of blocks
        self.progmean2=0.   #progressive squares mean of blocks
        self.blk_val_num=0. #number of values in a block
    
    def AppendVal(self,val):
        self.blk_sum+=val
        self.blk_val_num+=1.0
        if(self.blk_val_num==self.blk_size): #the block is full
            av=self.blk_sum/float(self.blk_size)
            self.progmean  = (self.progmean*self.blk_num+av)/(1.+self.blk_num)
            self.progmean2 = (self.progmean2*self.blk_num+av*av)/(1.+self.blk_num)
            self.blk_num+=1.0
            if(self.blk_num>1):
                self.err=np.sqrt((self.progmean2-(self.progmean**2.))/(self.blk_num-1))
            #reinitialization
            self.blk_sum=0.
            self.blk_val_num=0
            
    def flush(self):     #only keeps block size
        self.blk_sum=0.
        self.blk_num=0
        self.err=0.
        self.progmean=0.
        self.progmean2=0.
        self.blk_val_num=0.

            
#create an array of tools
tools = []
blocksizes = []
linsizes = np.linspace(10,5000,245)
for l in linsizes:
    blocksizes.append(int(l))

for s in blocksizes:
    t = DataBlockingTool(s)
    tools.append(t)
```

### Fase solida

```python
#load data and analyze
U, P = np.loadtxt("output/solid/solidraw.dat",unpack=True,usecols=(0,1))

Uerr = np.array([])
Perr = np.array([])

for t in tools:
    #Internal energy
    for u in U:
        t.AppendVal(u)
    Uerr=np.append(Uerr,t.err)
    #print(t.err,t.blk_size,t.blk_num,t.progmean)
    t.flush()
    #Pressure
    for p in P:
        t.AppendVal(p)
    Perr=np.append(Perr,t.err)
    #print(t.err,t.blk_num,t.progmean)
    t.flush()
    
plt.figure(figsize=(15,7))
plt.suptitle('Block sizes and errors - Solid phase',fontweight='bold')
plt.subplot(121)
plt.plot(blocksizes,Uerr)
plt.xlabel('block size')
plt.ylabel('error [$\sigma_{U/N}$]')
plt.title('Internal energy per particle errors')
plt.grid(True)
plt.subplot(122)
plt.plot(blocksizes,Perr)
plt.xlabel('block size')
plt.ylabel('error [$\sigma_P$]')
plt.title('Pressure errors')
plt.grid(True)
plt.show()
```

### Fase liquida

```python
#load data and analyze
U, P = np.loadtxt("output/liquid/liquidraw.dat",unpack=True,usecols=(0,1))

Uerr = np.array([])
Perr = np.array([])

for t in tools:
    #Internal energy
    for u in U:
        t.AppendVal(u)
    Uerr=np.append(Uerr,t.err)
    #print(t.err,t.blk_size,t.blk_num,t.progmean)
    t.flush()
    #Pressure
    for p in P:
        t.AppendVal(p)
    Perr=np.append(Perr,t.err)
    #print(t.err,t.blk_num,t.progmean)
    t.flush()
    
plt.figure(figsize=(15,7))
plt.suptitle('Block sizes and errors - Liquid phase',fontweight='bold')
plt.subplot(121)
plt.plot(blocksizes,Uerr)
plt.xlabel('block size')
plt.ylabel('error [$\sigma_{U/N}$]')
plt.title('Internal energy per particle errors')
plt.grid(True)
plt.subplot(122)
plt.plot(blocksizes,Perr)
plt.xlabel('block size')
plt.ylabel('error [$\sigma_P$]')
plt.title('Pressure errors')
plt.grid(True)
plt.show()
```

```python
#solid

from scipy.optimize import curve_fit

def expo(x,tc):
    return np.exp(-x/tc)

popt, pcov = curve_fit(expo,steps-eqsteps,ACU)

model=expo(steps-eqsteps,popt)

plt.figure(figsize=(15,6))
plt.plot(steps,ACU,label='data')
plt.plot(steps,model,label='fit')
plt.grid(True)
plt.xlabel('steps')
plt.ylabel('autocorrelation')
plt.title('Internal energy')
plt.legend()

print(popt)
```
