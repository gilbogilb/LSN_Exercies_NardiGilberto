# Esercizio 3 - Pricing di opzioni europee con metodi MonteCarlo
In questo esercizio si vogliono sfruttare tecniche MonteCarlo per calcolare il prezzo di call e di put di opzioni europee, sotto le ipotesi del modello Black-Scholes. Di tale modello si conosce la soluzione analitica, e si può di conseguenza controllare l'efficacia dei metodi utilizzati per confronto.

Questo tipo di opzioni sono concettualmente piuttosto semplici: il venditore dell'opzione garantisce al compratore la possibilità di comprare (_call option_) o vendere (_put option_) un asset per un prezzo prestabilito $K$ al tempo (futuro) $T$. Il profitto che il compratore ne ricava è dunque
$$
\max[0,S(T)-K]\equiv(S(T)-K)^+
$$
nel caso di opzioni call e 
$$
\max[0,K-S(T)]\equiv(S(T)-K)^+
$$
nel caso di opzioni put. Ovviamente il venditore richiederà un compenso in cambio dell'accordo: il nostro compito è proprio capire quanto dovrebbe costare l'opzione. Recuperiamo le ipotesi della teoria di Black-Scholes:
- Il mercato è perfettamente efficiente
- Il prezzo dell'asset evolve secondo un moto browniano geometrico ($\sim GBM(r,\sigma^2)$)

Se consideriamo una situazione descritta dai parametri:
- $S(t=0)=100$
- $T=1$
- $K=100$
- $r=0.1$
- $\sigma=0.25$

si ottiene analiticamente che i prezzi corretti al tempo $t=0$ sono **C=14.975** per opzioni **call** e **C=5.459** per opzioni **put**.

Possiamo in alternativa stimare il prezzo delle opzioni con dei metodi MonteCarlo. Questa soluzione è più generale e permette di prezzare opzioni diverse (o con ipotesi diverse) per le quali la soluzione analitica è sconosciuta. Nel caso di opzioni europee basta calcolare
$$
\langle e^{-rT}(S(T)-K)\rangle \quad \quad \langle e^{-rT}(K-S(T))\rangle
$$
. Nell'ipotesi in cui il prezzo dell'asset $S(T)$ evolva secondo un moto browniano geometrico, se ne può calcolare il valore ad uno step temporale successivo tramite la relazione ricorsiva

$$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\sigma^2/2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    con $Z_i \sim \mathcal{N}(0,1)$. Possiamo considerare due approcci diversi: evolvere direttamente il prezzo in un unico step temporale, oppure dividere il tempo tra $0$ e $T$ in una serie di step e calcolare di volta  in volta il nuovo prezzo fino a $t=T$.
    
Abbiamo sperimentato entrambi gli approcci: riportiamo i risultati ottenuti con medie su $10^4$ valori divisi in $100$ blocchi, evolvendo il prezzo dell'asset in un unico step oppure in $100$ step successivi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#data
call1, err_call1, calln, err_calln = np.loadtxt("data/EUpriceCALL.dat",unpack=True,usecols=(0,1,2,3),delimiter=' ')
put1, err_put1, putn, err_putn = np.loadtxt("data/EUpricePUT.dat",unpack=True,usecols=(0,1,2,3),delimiter=' ')
x=range(call1.size)

#analytical solutions
C = 14.975790778311286
P = 5.4595325819072364

Cvec = C*np.ones(call1.size)
Pvec = P*np.ones(put1.size)

#plotting
plt.figure(figsize=(16,12))
plt.suptitle('MonteCarlo European option pricing',fontweight='bold',fontsize=19)

plt.subplot(221)
plt.errorbar(x,call1,err_call1,label='MC pricing')
plt.title('Single evolution call price',fontsize=14)
plt.xlabel('blocks')
plt.ylabel('price')
plt.grid(True)
plt.plot(x,Cvec,'r--',label='Analytical solution')
plt.legend()

plt.subplot(222)
plt.errorbar(x,calln,err_calln,label='MC pricing')
plt.title('n=100 evolutions call price',fontsize=14)
plt.xlabel('blocks')
plt.ylabel('price')
plt.grid(True)
plt.plot(x,Cvec,'r--',label='Analytical solution')
plt.legend()

plt.subplot(223)
plt.errorbar(x,put1,err_put1,label='MC pricing')
plt.title('Single evolution put price',fontsize=14)
plt.xlabel('blocks')
plt.ylabel('price')
plt.grid(True)
plt.plot(x,Pvec,'r--',label='Analytical solution')
plt.legend()

plt.subplot(224)
plt.errorbar(x,putn,err_putn,label='MC pricing')
plt.title('n=100 evolutions put price',fontsize=14)
plt.xlabel('blocks')
plt.ylabel('price')
plt.grid(True)
plt.plot(x,Pvec,'r--',label='Analytical solution')
plt.legend()

plt.show()

print('---final results---')
print('single evolution')
print('call:',call1[call1.size-1],' +- ',err_call1[err_call1.size-1])
print('put:',put1[put1.size-1],' +- ',err_put1[err_put1.size-1])
print('n=100 evolutions')
print('call:',calln[calln.size-1],' +- ',err_calln[err_calln.size-1])
print('put:',putn[putn.size-1],' +- ',err_putn[err_putn.size-1])
```
